# Исследование объявлений о продаже квартир

Есть данные Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно выяснить рыночную стоимость объектов недвижимости. Необходимо установить параметры, выявить подозрительные объявления. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных:**
* airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
* balcony — число балконов
* ceiling_height — высота потолков (м)
* cityCenters_nearest — расстояние до центра города (м)
* days_exposition — сколько дней было размещено объявление (от публикации до снятия)
* first_day_exposition — дата публикации
* floor — этаж
* floors_total — всего этажей в доме
* is_apartment — апартаменты (булев тип)
* kitchen_area — площадь кухни в квадратных метрах (м²)
* last_price — цена на момент снятия с публикации
* living_area — жилая площадь в квадратных метрах (м²)
* locality_name — название населённого пункта
* open_plan — свободная планировка (булев тип)
* parks_around3000 — число парков в радиусе 3 км
* parks_nearest — расстояние до ближайшего парка (м)
* ponds_around3000 — число водоёмов в радиусе 3 км
* ponds_nearest — расстояние до ближайшего водоёма (м)
* rooms — число комнат
* studio — квартира-студия (булев тип)
* total_area — площадь квартиры в квадратных метрах (м²)
* total_images — число фотографий квартиры в объявлении

### Заполнение пропусков

#### Высота потолков. Столбец *ceiling_height*

```python
print(df['ceiling_height'].value_counts().sort_values(ascending=False).head(15)) # сортирую уникальные значения потолков по убыванию высоты
print(df['locality_name'].value_counts().head()) # количество уникальных названий населенных пунктов
df['ceiling_height'].hist(bins=100, range=(1, 4)); # гистограмма распределения высоты потолков
(
    display(df.query('locality_name == "Санкт-Петербург"') # высота потолков по Питеру
        .groupby('locality_name')
        .agg({'ceiling_height' : ['count', 'median', 'mean']}))
)
(
    df.query('city_centers_nearest < 8000 and locality_name == "Санкт-Петербург"') # высота потолков по центру Питера
    .groupby('locality_name')
    .agg({'ceiling_height' : ['count', 'median', 'mean']})
)
```

В столбце *ceiling_height* среднее не сильно отличается от медианы, 2.77 м. и 2.65 м. соответсвенно. 

На гистограмме видно, что в диапазоне от 2,5 до 3 м. типовые значения, их больше всего в выборке. Значения меньше 2.5 м. можно посчитать выбросом, так как по законодательству минимальная высота потолка жилых помещений допускается 2.5 м. Значения  4 м. и больше почти не встречаются в выборке, очень редкие, счет на единицы и десятки. Можно заменить медианой, сильных искажений не будет. К тому же высота потолков второстепенный фактор влияния на рыночную цену недвижимости.

Я заполню отдельно Санкт-Петербург, так как там наибольшая доля недвижимости в выборке - 66%. Его медиана 2.7 м. Однако, чем ближе к центру, тем больше высота потолков, чем дальше тем значение ближе к 2.7 м. На окраинах современные типовые застройки с высотой потолков по 2.5 м. в большинстве. При ограничении 8 км. от центра медиана высоты потолков в Санкт-Петербурге 3 м. Окраины Санкт-Петербурга я заполню его общей медианой в 2.7. Остальные населенные пункты общей медианой в 2.65 м.

```python
# заполнил временно пропуски -1
df['ceiling_height'] = df['ceiling_height'].fillna(-1) 
# заполнил по 2.7 м. дальше 8 км по Питеру
df.loc[(df['ceiling_height'] == -1) & 
       (df['city_centers_nearest'] > 8000) &
       (df['locality_name'] == 'Санкт-Петербург') |
       (df['city_centers_nearest'].isna()), 'ceiling_height'] = 2.7 
# заполнил по 3 м. в центре Питера
df.loc[(df['ceiling_height'] == -1) &
       (df['city_centers_nearest'] <= 8000) &
       (df['locality_name'] == 'Санкт-Петербург'), 'ceiling_height'] = 3.0 
# заполнил по 2.65 м. остальные, где есть значения в столбце locality_name
df.loc[(df['ceiling_height'] == -1) &
       (df['locality_name'] != 'Санкт-Петербург') |
       (df['locality_name'].isna()), 'ceiling_height'] = 2.65 
# заполнил по 2.65 м., где нет значений в locality_name
df.loc[(df['ceiling_height'] == -1) &
       (df['locality_name'].isna()), 'ceiling_height'] = 2.65
# заполнил по 2.7 м. для Питера, где пропуски в столбце расстояние до центра
df.loc[(df['ceiling_height'] == -1) & 
       (df['city_centers_nearest'].isna()) &
       (df['locality_name'] == 'Санкт-Петербург'), 'ceiling_height'] = 2.7
# проверил на пропуски
df[df['ceiling_height'] == -1]['ceiling_height'].count() 
```

В 613 строках мы знаем жилую и общую площади, но не знаем площадь кухни. Можно вычислить долю другой нежилой площади (туалеты, ванные и прочее). Вычтим жилую площадь из общей и домножим на разность 1 - средняя доля другой нежилой площади. Получим примерную площадь кухни. Вместо среднего значения нежилой доли возьмем медиану. Сначала найдем долю другой нежилой площади.

```python
# вычислил остальную нежилую площадь
areas_1['other_area'] = areas_1['total_area'] - (areas_1['kitchen_area'] + areas_1['living_area']) 
areas_1.query('other_area >= 0 and 75 < total_area <= 100').describe()
```

```python
df['kitchen_area'] = np.where((~(df['living_area'].isna())) & 
                              (df['kitchen_area'].isna()) &
                              (df['total_area'] > 27) &
                              (df['total_area'] <= 100), 
                              (df['total_area'] - df['living_area'])*0.755,df['kitchen_area'])
# посчитал, сколько осталось пропусков при известной жилой площади
print(df.query('kitchen_area.isna() and ~(living_area.isna())')['total_area'].count()) 
# посчитал, сколько всего пропусков по кухням
df['kitchen_area'].isna().sum() 
```

Мы можем судить по данным за 2017 и 2018 года. Всплеск продаж недвижимости в диапазоне 45 - 50 дней связан с тем, что по медиане цена за кв. м. меньше, чем по рынку. За 2018 г. 78 000 против 96 000 по рынку. Такой всплеск продаж выглядит правдоподобно (демпинг), в отличие от всплеска в диапазоне 5-10 дней, когда цена за кв. м. была сравнима с общей. Скорее всего, это показатель не продажи недвижимости, а снятия объявления. Посмотрим подробней на этот всплеск относительно цен за первые 30 дней 2017-2018 года

```python
(df.query('days_exposition <=30 and year in [2017, 2018]')
 .pivot_table(index='days_exposition_category', values='price_per_sq_meter', aggfunc=['count', 'mean', 'median'])
)
```
